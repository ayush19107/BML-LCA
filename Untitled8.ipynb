```python
# ---------------- BASIC EDA IN ONE BLOCK ----------------

import pandas as pd
import zipfile
import io
import requests
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load dataset directly from UCI ZIP
zip_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00222/bank.zip"
response = requests.get(zip_url)
z = zipfile.ZipFile(io.BytesIO(response.content))
df = pd.read_csv(z.open('bank.csv'), sep=';')

# 1. Basic info
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())

# 2. Summary stats
print("\nSummary Statistics:")
print(df.describe())

# 3. Missing values
print("\nMissing Values:")
print(df.isnull().sum())

# 4. Target distribution
sns.countplot(x=df['y'])
plt.title("Target Distribution (Subscribed or Not)")
plt.show()

print("\nTarget Value Counts:")
print(df['y'].value_counts())

# 5. Correlation matrix (numeric only)
num_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(8,6))
sns.heatmap(df[num_cols].corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

```python
# ---- STEP 3: Select Features (job, duration, age, balance) ----

features = ['job', 'duration', 'age', 'balance']
X = pd.get_dummies(df[features], drop_first=True)   # one-hot encode job
y = df['y'].map({'yes':1, 'no':0})

X.head(), y.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.30, random_state=42, stratify=y
)

scaler = StandardScaler()
X_train[['age','balance']] = scaler.fit_transform(X_train[['age','balance']])
X_test[['age','balance']] = scaler.transform(X_test[['age','balance']])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score

model = LogisticRegression(class_weight='balanced', max_iter=1000)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
y_prob = model.predict_proba(X_test)[:,1]
```

```python
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))

print("Classification Report:")
print(classification_report(y_test, y_pred))

print("ROC-AUC:", roc_auc_score(y_test, y_prob))
```
